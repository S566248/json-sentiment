```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load spaCy model and add SpacyTextBlob to the pipeline
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')

# Read in the contents of the JSON file
with open('lyrics.json', 'r') as f:
    data = json.load(f)

# Extract the lyrics from the JSON data
lyrics = data['lyrics']

# Print the lyrics of the song
print("Lyrics of the song:")
print(lyrics)
print()

# Analyze sentiment using spaCyTextBlob
doc = nlp(lyrics)

# Print the polarity score of the sentiment analysis
print("Polarity score:", doc._.polarity)

# Based on the polarity score, determine the connotation of the lyrics
# Comment: 
# A polarity score closer to 1.0 indicates a positive sentiment,
# while a polarity score closer to -1.0 indicates a negative sentiment.
# Therefore, if the polarity score is positive, the lyrics have a more positive connotation,
# and if the polarity score is negative, the lyrics have a more negative connotation.
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import requests
import json

def get_lyrics(artist, song, filename):
    """
    Get the lyrics of a song by the given artist and store them in a file.

    Args:
    - artist (str): The name of the artist.
    - song (str): The title of the song.
    - filename (str): The filename to store the lyrics.

    Returns:
    - None
    """
    # Access the API and load the JSON response into a dictionary
    result = json.loads(requests.get(f'https://api.lyrics.ovh/v1/{artist}/{song}').text)

    # Write the lyrics to a file
    with open(filename, 'w') as f:
        if 'lyrics' in result:
            f.write(result['lyrics'])
        else:
            f.write("Lyrics not found.")

# Test the function by getting the lyrics of four songs and storing them in different files
get_lyrics("Owl City", "Fireflies", "fireflies_lyrics.txt")
get_lyrics("Kelly Clarkson", "Already Gone", "already_gone_lyrics.txt")
get_lyrics("Hot Chelle Rae", "Tonight Tonight", "tonight_tonight_lyrics.txt")
get_lyrics("Adele", "Someone Like You", "someone_like_you_lyrics.txt")
```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

def get_polarity_score(filename):
    """
    Load lyrics from a file, perform sentiment analysis, and return the polarity score.

    Args:
    - filename (str): The name of the file containing the song lyrics.

    Returns:
    - float: The polarity score of the sentiment analysis.
    """
    # Load spaCy model and add SpacyTextBlob to the pipeline
    nlp = spacy.load('en_core_web_sm')
    nlp.add_pipe('spacytextblob')

    # Read the lyrics from the file
    with open(filename, 'r') as f:
        lyrics = f.read()

    # Analyze sentiment using spaCyTextBlob
    doc = nlp(lyrics)

    # Return the polarity score
    return doc._.polarity

# Test the function by printing the polarity scores of the three files created in question 3
print("Polarity score for Fireflies:", get_polarity_score("fireflies_lyrics.txt"))
print("Polarity score for Already Gone:", get_polarity_score("already_gone_lyrics.txt"))
print("Polarity score for Someone Like You:", get_polarity_score("someone_like_you_lyrics.txt"))
print("Polarity score for Tonight Tonight:", get_polarity_score("tonight_tonight_lyrics.txt"))
```

Whether the reported polarity matches our understanding of the song's lyrics depends on various factors, including the complexity of the lyrics, the context of the song, and individual interpretation.
